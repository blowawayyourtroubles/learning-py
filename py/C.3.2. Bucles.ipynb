```python
# L√ìGICA CONDICIONAL Y BUCLES EN PYTHON
print("=== CONDICIONALES IF-ELSE Y BUCLES FOR-WHILE ===\n")

# 1. CONDICIONAL IF B√ÅSICO
print("1. CONDICIONAL IF B√ÅSICO:")
edad = 18
print(f"Edad: {edad}")

if edad >= 18:
    print("‚úÖ Eres mayor de edad")
    print("Puedes votar en las elecciones")

print("\n" + "="*50)
```

```python
# 2. IF-ELSE
print("\n2. CONDICIONAL IF-ELSE:")
temperatura = 25
print(f"Temperatura: {temperatura}¬∞C")

if temperatura > 30:
    print("üå°Ô∏è Hace mucho calor")
else:
    print("üå§Ô∏è La temperatura es agradable")

print("\n" + "="*50)
```

```python
# 3. IF-ELIF-ELSE (M√∫ltiples condiciones)
print("\n3. CONDICIONAL IF-ELIF-ELSE:")
calificacion = 85
print(f"Calificaci√≥n: {calificacion}")

if calificacion >= 90:
    print("üéâ Excelente - Calificaci√≥n A")
elif calificacion >= 80:
    print("üëç Muy bien - Calificaci√≥n B")
elif calificacion >= 70:
    print("‚úÖ Bien - Calificaci√≥n C")
elif calificacion >= 60:
    print("‚ö†Ô∏è Aprobado - Calificaci√≥n D")
else:
    print("‚ùå Reprobado - Calificaci√≥n F")

print("\n" + "="*50)
```

```python
# 4. BUCLE FOR con lista
print("\n4. BUCLE FOR CON LISTA:")
frutas = ["manzana", "banana", "naranja", "uva"]
print("Lista de frutas:", frutas)

print("Recorriendo la lista:")
for fruta in frutas:
    print(f"  - Me gusta la {fruta}")

print("\n" + "="*50)
```

```python

# 5. BUCLE FOR con range()
print("\n5. BUCLE FOR CON RANGE():")
print("Contando del 1 al 5:")
for i in range(1, 6):
    print(f"  N√∫mero: {i}")

print("\nTabla del 3:")
for i in range(1, 11):
    resultado = 3 * i
    print(f"  3 √ó {i} = {resultado}")

print("\n" + "="*50)
```

```python
# 6. BUCLE WHILE b√°sico
print("\n6. BUCLE WHILE B√ÅSICO:")
contador = 1
print("Contando hasta 5 con while:")
while contador <= 5:
    print(f"  Contador: {contador}")
    contador += 1  # Incrementar el contador

print("\n" + "="*50)
```

```python
# 7. BUCLE WHILE con condici√≥n compleja
print("\n7. BUCLE WHILE CON CONDICI√ìN COMPLEJA:")
numero_secreto = input ("Ingresa un n√∫mero secreto: ")
print(f"N√∫mero secreto: {numero_secreto}")
intento = 0
max_intentos = 3

print("Adivina el n√∫mero secreto (1-10)")
print(f"Tienes {max_intentos} intentos")

while intento < max_intentos:
    guess = int(input("Ingresa tu n√∫mero: "))

    if guess == numero_secreto:
        print("üéØ ¬°Correcto! Adivinaste el n√∫mero")
        break  # Salir del bucle
    elif guess < numero_secreto:
        print("üìà El n√∫mero es mayor")
    else:
        print("üìâ El n√∫mero es menor")

    intento += 1
    intentos_restantes = max_intentos - intento
    print(f"Intentos restantes: {intentos_restantes}")

if intento == max_intentos and guess != numero_secreto:
    print(f"‚ùå Game Over. El n√∫mero era {numero_secreto}")

print("\n" + "="*50)
```

```python
# 8. COMBINACI√ìN: Bucles + Condicionales
print("\n8. COMBINACI√ìN BUCLES + CONDICIONALES:")
numeros = [12, 5, 8, 19, 3, 15, 7, 20]
print(f"Lista de n√∫meros: {numeros}")

print("N√∫meros pares mayores a 10:")
for numero in numeros:
    if numero > 10 and numero % 2 == 0:
        print(f"  {numero} es par y mayor que 10")

print("\n" + "="*50)
```

```python
# 9. CONTROL DE BUCLES: break, continue
print("\n9. CONTROL DE BUCLES: BREAK Y CONTINUE:")
print("N√∫meros del 1 al 10 (saltando el 5 y parando en 8):")
for i in range(1, 11):
    if i == 5:
        continue  # Saltar esta iteraci√≥n
    if i > 8:
        break     # Salir del bucle
    print(f"  N√∫mero: {i}")

print("\n¬°Programa terminado! üêç")
```
