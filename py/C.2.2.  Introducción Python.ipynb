```python
# OPERACIONES ARITMÉTICAS EN PYTHON - EJEMPLO PRÁCTICO
print("=== CALCULADORA BÁSICA CON OPERADORES ARITMÉTICOS ===\n")

# 1. Operaciones básicas con números enteros
print("1. OPERACIONES BÁSICAS:")
print("Suma (5 + 3):", 5 + 3)
print("Resta (10 - 4):", 10 - 4)
print("Multiplicación (6 * 7):", 6 * 7)
print("División (15 / 2):", 15 / 2)
print("División entera (15 // 2):", 15 // 2)
print("Módulo/residuo (15 % 2):", 15 % 2)
print("Potencia (2 ** 5):", 2 ** 5)

```

```python
# 2. Jerarquía de operaciones
print("\n2. JERARQUÍA DE OPERACIONES:")
print("Sin paréntesis (3 + 5 * 2):", 3 + 5 * 2)
print("Con paréntesis ((3 + 5) * 2):", (3 + 5) * 2)
```

```python
# 3. Operaciones con decimales
print("\n3. OPERACIONES CON DECIMALES:")
print("Suma decimal (2.5 + 3.7):", 2.5 + 3.7)
print("Multiplicación decimal (1.5 * 4):", 1.5 * 4)
```

```python
# 4. Operaciones combinadas
print("\n4. OPERACIÓN COMBINADA:")
resultado = (10 + 5) * 2 ** 3 / 4 - 2
print("(10 + 5) * 2 ** 3 / 4 - 2 =", resultado)
```

```python
# 5. Trabajando con variables
print("\n5. USO DE VARIABLES:")
a = 20
b = 6
print(f"a = {a}, b = {b}")
print("a + b =", a + b)
print("a * b - 10 =", a * b - 10)

print("\n¡Esto es todo! :)")
```
