```python
# MANIPULACIÃ“N DE LISTAS EN PYTHON
print("=== OPERACIONES CON LISTAS: APPEND, INSERT, REMOVE ===\n")

# 1. CreaciÃ³n de listas y mostrar contenido
print("1. CREACIÃ“N DE LISTAS:")
frutas = ["manzana", "banana", "naranja"]
numeros = [1, 2, 3, 4, 5]
mixta = ["texto", 100, True, 3.14]

print("Lista de frutas:", frutas)
print("Lista de nÃºmeros:", numeros)
print("Lista mixta:", mixta)
print(f"Longitud de frutas: {len(frutas)} elementos")
```

```python
# 2. MÃ©todo APPEND - Agregar al final
print("\n2. MÃ‰TODO APPEND (agregar al final):")
print("Lista original:", frutas)
frutas.append("uva")
print("DespuÃ©s de append('uva'):", frutas)
frutas.append("sandÃ­a")
print("DespuÃ©s de append('sandÃ­a'):", frutas)
```

```python
# 3. MÃ©todo INSERT - Insertar en posiciÃ³n especÃ­fica
print("\n3. MÃ‰TODO INSERT (insertar en posiciÃ³n):")
print("Lista actual:", frutas)
frutas.insert(1, "fresa")  # Inserta en posiciÃ³n 1
print("DespuÃ©s de insert(1, 'fresa'):", frutas)
frutas.insert(0, "kiwi")   # Inserta al inicio (posiciÃ³n 0)
print("DespuÃ©s de insert(0, 'kiwi'):", frutas)
frutas.insert(3, "mango")  # Inserta en posiciÃ³n 3
print("DespuÃ©s de insert(3, 'mango'):", frutas)
```

```python
# 4. MÃ©todo REMOVE - Eliminar elemento especÃ­fico
print("\n4. MÃ‰TODO REMOVE (eliminar elemento):")
print("Lista actual:", frutas)
frutas.remove("banana")
print("DespuÃ©s de remove('banana'):", frutas)
frutas.remove("naranja")
print("DespuÃ©s de remove('naranja'):", frutas)
```

```python

# 5. Manejo de errores con REMOVE
print("\n5. MANEJO DE ERRORES CON REMOVE:")
print("Lista actual:", frutas)
try:
    frutas.remove("pera")  # Elemento que no existe
    print("DespuÃ©s de remove('pera'):", frutas)
except ValueError as e:
    print(f"Error al remover 'pera': {e}")
```

```python
# 6. Otros mÃ©todos Ãºtiles de listas
print("\n6. OTROS MÃ‰TODOS ÃšTILES:")
# Re-initialize the 'frutas' list for this cell
frutas = ['kiwi', 'manzana', 'fresa', 'mango', 'uva', 'sandÃ­a']

# POP - elimina y retorna elemento de una posiciÃ³n
elemento_eliminado = frutas.pop(2)
print(f"pop(2) eliminÃ³: '{elemento_eliminado}'")
print("Lista despuÃ©s de pop(2):", frutas)

# INDEX - encuentra posiciÃ³n de un elemento
posicion = frutas.index("mango")
print(f"index('mango') = {posicion}")

# COUNT - cuenta ocurrencias
frutas.append("manzana")
print("DespuÃ©s de append('manzana'):", frutas)
conteo = frutas.count("manzana")
print(f"count('manzana') = {conteo}")
```

```python
# 7. Ejemplo prÃ¡ctico completo
print("\n7. EJEMPLO PRÃCTICO: LISTA DE COMPRAS")
compras = ["pan", "leche", "huevos"]
print("Lista de compras inicial:", compras)

# Agregar elementos
compras.append("queso")
compras.insert(1, "frutas")
print("DespuÃ©s de agregar elementos:", compras)

# Modificar la lista
compras.remove("huevos")
compras.append("yogurt")
print("DespuÃ©s de modificar:", compras)

# Mostrar lista final
print("\n=== LISTA FINAL ===")
for i, producto in enumerate(compras, 1):
    print(f"{i}. {producto}")

print(f"\nTotal de productos: {len(compras)}")
print("Â¡Lista completa! ðŸ›’")
```
